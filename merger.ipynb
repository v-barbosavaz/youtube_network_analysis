```python
import pandas as pd
from functools import reduce
from os import walk
import numpy as np
```

```python
f = []
for (dirpath, dirnames, filenames) in walk("data/generated/nodes_with_cat"):
    for file in filenames:       
        if file.endswith("nodes.csv"):
            f.append(file)
f
```
